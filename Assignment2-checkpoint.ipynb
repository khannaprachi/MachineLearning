```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 

bureaudata = pd.read_csv("C:\\Users\\khann\\Desktop\\Python Assignments\\home-credit-default-risk\\bureau.csv")
applicationdata = pd.read_csv("C:\\Users\\khann\\Desktop\\Python Assignments\\home-credit-default-risk\\application_train.csv")
creditcardbalance = pd.read_csv("C:\\Users\\khann\\Desktop\\Python Assignments\\home-credit-default-risk\\credit_card_balance.csv")
activecredits=bureaudata.groupby('CREDIT_ACTIVE').get_group('Active')
bureaucreditdata = activecredits.groupby(['SK_ID_CURR'], as_index=False)['AMT_CREDIT_SUM','AMT_CREDIT_SUM_DEBT','AMT_CREDIT_SUM_OVERDUE'].sum()
app_bureau_data=pd.merge(applicationdata,bureaucreditdata,on='SK_ID_CURR')
app_bureau_data.head()
creditcarddata = creditcardbalance.groupby(['SK_ID_CURR'], as_index=False)['AMT_BALANCE','AMT_INST_MIN_REGULARITY'].sum()
app_bureau_credit_data = pd.merge(app_bureau_data,creditcarddata,on='SK_ID_CURR')
app_bureau_credit_data.index
```

```python
X = app_bureau_credit_data.iloc[:,[7,8,17,18,122,123,124,125,126]].values.astype(int) 
Y = app_bureau_credit_data.iloc[:,1].values.astype(int)
print(X)
print(Y)

from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split
#regressor = DecisionTreeRegressor(random_state = 0,max_depth = 3)
#X= X.reshape(-1, 1)
X_train, X_test, y_train, y_test=train_test_split(X, Y, test_size=20000, train_size=100000, random_state=0, shuffle=False, stratify=None)
#X_train, X_test, y_train, y_test=train_test_split(X, Y,shuffle=False, random_state=0)
clf = DecisionTreeClassifier(random_state = 0)
y_train=y_train.reshape(-1,1)
print(y_train)
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict([[100000,650000]])
print("Predicted credit: % d\n"% y_pred)
from sklearn import tree
plt.figure(figsize=(80,15))
#tree.plot_tree(clf,class_names=True)
#tree.plot_tree(clf,filled=True,rounded=True,class_names=["Fault","Good"],feature_names=X)
tree.plot_tree(clf,filled=True,rounded=True,class_names=True,feature_names=X_train)
plt.show()
```

```python
print(clf.score(X_train, y_train))
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(clf, X_test, y_test)
plt.show()
```

```python
path = clf.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities
ccp_alphas=ccp_alphas[:-1]
#print(ccp_alphas)
clf_dts =[]
ccp_alphastotal=[]

for ccp_alphatest in ccp_alphas:
    if ccp_alphatest >= 0.0 : 
        ccp_alphastotal.append(ccp_alphatest)
        clf_dt=DecisionTreeClassifier(random_state = 0,ccp_alpha=ccp_alphatest)
        clf_dt.fit(X_train,y_train)
        clf_dts.append(clf_dt)

train_scores = [clf.score(X_train, y_train) for clf in clf_dts]
test_scores = [clf.score(X_test, y_test) for clf in clf_dts]

fig, ax = plt.subplots(figsize=(10, 15))
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax.plot(ccp_alphastotal, train_scores, marker="o", label="train", drawstyle="steps-post")
ax.plot(ccp_alphastotal, test_scores, marker="o", label="test", drawstyle="steps-post")
plt.xlim(0, .0005);
#plt.ylim(-1, 2);
ax.legend()
plt.show()
    
```
