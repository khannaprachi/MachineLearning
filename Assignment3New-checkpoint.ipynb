```python
def create_importance_dataframe(pca, applicationdata):

    # Change pcs components ndarray to a dataframe
    importance_df  = pd.DataFrame(pca.components_)

    # Assign columns
    importance_df.columns  = applicationdata.columns

    # Change to absolute values
    importance_df =importance_df.apply(np.abs)

    # Transpose
    importance_df=importance_df.transpose()

    # Change column names again

    ## First get number of pcs
    num_pcs = importance_df.shape[1]

    ## Generate the new column names
    new_columns = [f'PC{i}' for i in range(1, num_pcs + 1)]

    ## Now rename
    importance_df.columns  =new_columns

    # Return importance df
    return importance_df



# Call function to create importance df
importance_df  =create_importance_dataframe(pca_out, applicationdata)

# Show first few rows
display(importance_df)

# Sort depending on PC of interest

## PC1 top 10 important features
pc1_top_10_features = importance_df['PC1'].sort_values(ascending = False)[:10]
print(), print(f'PC1 top 10 feautres are \n')
display(pc1_top_10_features )

## PC2 top 10 important features
pc2_top_10_features = importance_df['PC2'].sort_values(ascending = False)[:10]
print(), print(f'PC2 top 10 feautres are \n')
display(pc2_top_10_features )
## PC2 top 10 important features
pc2_top_10_features = importance_df['PC3'].sort_values(ascending = False)[:10]
print(), print(f'PC3 top 10 feautres are \n')
display(pc2_top_10_features )
## PC2 top 10 important features
pc2_top_10_features = importance_df['PC4'].sort_values(ascending = False)[:10]
print(), print(f'PC4 top 10 feautres are \n')
display(pc2_top_10_features )
## PC2 top 10 important features
pc2_top_10_features = importance_df['PC5'].sort_values(ascending = False)[:10]
print(), print(f'PC5 top 10 feautres are \n')
display(pc2_top_10_features )
## PC2 top 10 important features
pc2_top_10_features = importance_df['PC6'].sort_values(ascending = False)[:10]
print(), print(f'PC6 top 10 feautres are \n')
display(pc2_top_10_features )
## PC2 top 10 important features
pc2_top_10_features = importance_df['PC7'].sort_values(ascending = False)[:10]
print(), print(f'PC7 top 10 feautres are \n')
display(pc2_top_10_features )
## PC2 top 10 important features
pc2_top_10_features = importance_df['PC8'].sort_values(ascending = False)[:10]
print(), print(f'PC8 top 10 feautres are \n')
display(pc2_top_10_features )
```

```python
# Section 2 
import scipy.cluster.hierarchy as hc
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

# Change to absolute values
df_selection=pd.DataFrame(pca_out.components_)
df_selection=df_selection.apply(np.abs)
# Transpose
df_selection=df_selection.transpose()
    
df_pca = pd.DataFrame(df_selection)
df_pca=df_pca.iloc[:, 0:3]
df_pca.columns = ['P1', 'P2', 'P3']
print(df_pca)
```

```python
# Create the Dendogram plot
#
plt.figure(figsize =(8, 8))
plt.title('Visualising the data')
dendrogram = hc.dendrogram((hc.linkage(df_pca, method ='ward')))
```

```python
#
# Create the clusters using Agglomerative hierarchical clustering
#



#fig = plt.figure(figsize=(20, 8))
#plt.subplot(5, 2, 1)
agc = AgglomerativeClustering(n_clusters = 10)
plt.figure(figsize =(4, 4))
plt.scatter(df_pca['P1'], df_pca['P2'], c = agc.fit_predict(df_pca), cmap ='rainbow')
plt.show()

#plt.subplot(5, 2, 2)
agc = AgglomerativeClustering(n_clusters = 8)
plt.figure(figsize =(4, 4))
plt.scatter(df_pca['P1'], df_pca['P2'], c = agc.fit_predict(df_pca), cmap ='rainbow')
plt.show()

#plt.subplot(5, 2, 3)
agc = AgglomerativeClustering(n_clusters = 5)
plt.figure(figsize =(4, 4))
plt.scatter(df_pca['P1'], df_pca['P2'], c = agc.fit_predict(df_pca), cmap ='rainbow')
plt.show()

#plt.subplot(5, 2, 4)
agc = AgglomerativeClustering(n_clusters = 4)
plt.figure(figsize =(4, 4))
plt.scatter(df_pca['P1'], df_pca['P2'], c = agc.fit_predict(df_pca), cmap ='rainbow')
plt.show()

#plt.subplot(5, 2, 5)
agc = AgglomerativeClustering(n_clusters = 3)
plt.figure(figsize =(4, 4))
plt.scatter(df_pca['P1'], df_pca['P2'], c = agc.fit_predict(df_pca), cmap ='rainbow')
plt.show()

#plt.subplot(5, 2, 6)
agc = AgglomerativeClustering(n_clusters = 2)
plt.figure(figsize =(4, 4))
plt.scatter(df_pca['P1'], df_pca['P2'], c = agc.fit_predict(df_pca), cmap ='rainbow')
plt.title("Agglomerative Hierarchical Clusters - Scatter Plot", fontsize=18)
plt.show()
```

```python
from sklearn.cluster import KMeans

plt.scatter(df_pca['P1'], df_pca['P2'],
             color='white', marker='o', edgecolor='red', s=50)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
wcss=[]
for i in range(1,11):
    
    kmeans = KMeans(n_clusters=i, init ='k-means++', max_iter=300,  n_init=10,random_state=0)
    kmeans.fit(df_pca)
    wcss.append(kmeans.inertia_)
    
print(wcss)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

kmeans = KMeans(n_clusters=4, init ='k-means++', max_iter=300, n_init=10,random_state=0 )
y_kmeans = kmeans.fit_predict(df_pca)
y_kmeans
```

```python
plt.scatter(df_pca.iloc[y_kmeans==0, 0], df_pca.iloc[y_kmeans==0, 1], s=100, c='red', label ='Cluster 1')
plt.scatter(df_pca.iloc[y_kmeans==1, 0], df_pca.iloc[y_kmeans==1, 1], s=100, c='blue', label ='Cluster 2')
plt.scatter(df_pca.iloc[y_kmeans==2, 0], df_pca.iloc[y_kmeans==2, 1], s=100, c='green', label ='Cluster 3')
plt.scatter(df_pca.iloc[y_kmeans==3, 0], df_pca.iloc[y_kmeans==3, 1], s=100, c='cyan', label ='Cluster 4')
#plt.scatter(df_pca.iloc[y_kmeans==4, 0], df_pca.iloc[y_kmeans==4, 1], s=100, c='magenta', label ='Cluster 5')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label = 'Centroids')
plt.title('Clusters of Data')
plt.xlabel('PC2')
plt.ylabel('PC1')
plt.show()
```
